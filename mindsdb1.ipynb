```python
import base64
import requests
import pandas as pd
import yfinance as yf 
```

```python
msg = "openbb:mindsdb2024"
msg_bytes = msg.encode("ascii")
base64_bytes = base64.b64encode(msg_bytes)
base64_msg = base64_bytes.decode("ascii")

def callandstore(parameter_list):
    
    json_results = []
    df = None
    
    for value in symbol_list:
        # Construct the dynamic API request URL with the current value
        url = f"http://mindsdb2024.openbb.dev/api/v1/equity/price/quote?provider=intrinio&symbol={value}&source=intrinio_mx"
        headers = {"accept": "application/json", "Authorization": f"Basic {base64_msg}"}
        response = requests.get(url=url, headers=headers)
        #print(response)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Append the JSON result to the list
            json_results.append(response.json())     
            #print(response.status_code)

            # Convert the list of JSON results to a DataFrame
            if df is not None:
                temp_df = pd.DataFrame(response.json()['results'])
                df = pd.concat([df, temp_df], ignore_index=True)
                #print("went to loop1")
            else:
                df = pd.DataFrame(response.json()['results'])
                #print("went to loop2")
            
        else:
            # Print an error message if the request was not successful
            print(f"Error for value {value}: {response.status_code}")
    return df
```
