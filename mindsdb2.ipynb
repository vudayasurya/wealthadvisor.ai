```python
def callandstore(parameter_list):
    
    json_results = []
    df = None
    
    for value in parameter_list:
        # Construct the dynamic API request URL with the current value
        API_KEY = '41f219c66feb48f1b53db8a67016c4ac'
        url = f"https://newsapi.org/v2/top-headlines?q={value}&country=us&source=google_news&apiKey={API_KEY}"
        response = requests.get(url=url)
        #print(response)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Append the JSON result to the list
            json_results.append(response.json())     
            #print(response.status_code)
            
            index = 0
            
            # Convert the list of JSON results to a DataFrame
            while index < len(response.json()['articles']):
                if df is not None:
                    temp_df['title'] = pd.DataFrame(response.json()['articles'][index]['title'])
                    df = pd.concat([df, temp_df], ignore_index=True)
                #print("went to loop1")
                else:
                    df = pd.DataFrame(response.json()['articles'][index]['title'])
                #print("went to loop2")
                
            index += 1
                
        else:
            # Print an error message if the request was not successful
            print(f"Error for value {value}: {response.status_code}")
    return df
```
